Practice

```python
from functools import reduce

with open("./sample_data/sample.txt", "r") as file:
  # sentences = file.read()
  # print(sentences)

  # Number of lines
  lst = file.readlines()
  num_lines = len(lst)
  print(num_lines)

  # Number of words
  num_words = list(map(lambda x : len(x.split()), lst))
  print(num_words)
  total_words = reduce(lambda x, y : x + y, num_words)
  print(total_words)
```

Exception Handling

```python
# a = 10 / 0 # Error
# print("Hello World") # Not error- But why showing error?

num = int(input())

try:
  a = 10 / num
except ZeroDivisionError:
  print("You can't divide a number by zero")

print(num)
```

```python
num = int(input())

try:
  a = 10 / num
except Exception as e:
  print("Error:", e)

print(num)
```
