### Tuples

```python
tup = (10, 20, 30, 40, 50)
# print(type(tup))

mixed_tup = (10, 10.5, "hello", True)
print(mixed_tup)
```

```python
# list is mutable
list = [10, 20, 30]
list[1] = 15
print(list)

# But tuples are immutable
tup = (20, 30, 10)
# tup[1] = 15 // Error
print(tup)
```

```python
tup = (10, 10, 20, 20, 20, 30, 40, 40, 50)

print(tup.count(10))
print(tup.count(20))

print (tup.index(20)) # First index of 20 
```

```python
# A = {1, 2, 3}
# print(type(A), A)

# S = {1, 2, 2, 3, 3, 3} # Unique values
# print(S)

S = {5, 3, 2, 6, 7, 1} 
print(S) # Sorted

sum = 0

for element in S:
    # print(element)
    sum += element
print(sum)
```

```python
S = {1, 5, 4, 6, 7} # Immtable

S.add(2)
S.add(3)
print(S)

S.pop()
print(S)

S.pop()
print(S)

S.remove(5)
print(S)
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.union(b))
print(a.intersection(b))
print(a.isdisjoint(b))
print(a.issubset(b))
```

### Dictionaries

```python
dic = {"name" : "Tanvir", "age" : 20, "address" : "Dhaka", "numbers" : [10, 20, 30]}
# print(type(dic))
print(dic)

print(dic["name"])
print(dic["age"])
print(dic["address"])
print(dic["numbers"])

dic = {"name" : "Tanvir", "age" : 20, "address" : "Dhaka", "numbers" : [10, 20, 30], "age": 30} # mutable

print(dic["name"])
print(dic["age"])
print(dic["address"])
print(dic["numbers"])
```

```python
print(dic.get("marks"))
dic["marks"] = 80
print(dic)

dic.update({"English marks": 87})
print(dic)

```

```python
keys = dic.keys()
print(keys)
```
