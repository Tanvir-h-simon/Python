Data Distribution: A data distribution describes how values are spread out in a dataset. Example:
If we collect 10,000 students’ heights and plot them, the shape will be a bell curve, that is the distribution of the data.

Random Distribution: A random distribution is a distribution generated by a probability model. Instead of observing real-world data, we create data artificially according to a known mathematical rule.
Instead of observing real-world data, we create data artificially according to a known mathematical rule.
Examples:

normal() → generates values based on a normal (Gaussian) distribution

uniform() → generates values between low and high with equal probability

binomial() → simulates coin flips

poisson() → simulates event counts

A random distribution is a set of random numbers that follow a certain probability density function. Probability Density Function: A function that describes a continuous probability. i.e. probability of all values in an array.

Random Permutations: A permutation refers to an arrangement of elements. 

Shuffling Arrays:

```python
from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr) 
print(arr) # The shuffle() method makes changes to the original array.
```

Permutation of arrays:

```python
# from numpy import random
# import numpy as np
import numpy as np
np.random

arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr)) # The permutation() method returns a re-arranged array and leaves the original array un-changed.
```
