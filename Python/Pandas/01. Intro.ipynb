Pandas is a Python library used for working with data sets. It has functions for analyzing, cleaning, exploring, and manipulating data.

```python
import pandas

print(pandas.__version__)
```

Series: A Pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.

```python
arr = [1, 2, 3, 4, 5]
dataset = pandas.Series(arr)
print(dataset)
print()

print(dataset.shape, "\n")

print(dataset[0], "\n")
print(dataset[1:4], "\n")
```

Dataframe: A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns. 

```python
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

dataset = pandas.DataFrame(data)
print(dataset)
print()

print(dataset.shape, "\n")

print(dataset["calories"], "\n")
print(dataset["duration"], "\n")

# Pandas use the loc attribute to return one or more specified row(s)
print(dataset.loc[0], "\n")
print(dataset.loc[1:2], "\n")
```
