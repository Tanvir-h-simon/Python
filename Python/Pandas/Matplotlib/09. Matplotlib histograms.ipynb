Histogram: A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.

Create Histogram: The hist() function to create histograms. The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
import numpy as np

x = np.random.normal(170, 10, 250) # # 250 random integers between 170 and 10
print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random data between 1 and 20 
data = np.random.randint(1, 20, 50)

plt.hist(data, bins = 5, color = 'skyblue', edgecolor = 'black') # plt.hist(data, bins=number_of_bins(number of intervals to divide your data into), color='color', edgecolor='color')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
