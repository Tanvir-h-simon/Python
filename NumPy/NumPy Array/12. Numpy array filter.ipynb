Boolean Indexing- A boolean index list is a list of booleans corresponding to indexes in the array. If the value at an index is True that element is contained in the filtered array, if the value at that index is False that element is excluded from the filtered array.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
filtered = arr > 3
print(filtered)
print(arr[filtered])
```

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
filter_arr = arr % 2 == 0
new_arr = arr[filter_arr]
print(filter_arr)
print(new_arr)
```

```python
arr = np.array([5, 10, 15, 20, 25])

filtered = arr[(arr > 10) & (arr < 25)]
print(filtered)
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

filtered = arr[arr > 3]
print(filtered)
```

extract()- Equivalent to boolean indexing

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
filtered = np.extract(arr > 2, arr)
print(filtered)
```

where()- can return indices that satisfy a condition

```python
import numpy as np

arr = np.array([10, 20, 15, 30])
indices = np.where(arr > 15)
filtered = arr[indices]
print(filtered)
```
