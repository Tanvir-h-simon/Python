What is a Random Number?

Random number does NOT mean a different number every time. Random means something that can not be predicted logically.

Random numbers generated through a generation algorithm are called pseudo random.

Generate random float:

```python
import numpy as np

np.random.rand()  # Generate a random float number between 0 and 1
```

```python
np.random.rand(5)  # 1D array of 5 floats (between 0 and 1)
```

```python
np.random.rand(2, 3)  # 2x3 array
```

```python
import numpy as np

random_number_generator = np.random.default_rng()

random_number_generator.random()
random_number_generator.random(5)
random_number_generator.random((2, 3))
```

```python
from numpy import random

num = random.rand() # Generate a random float number between 0 and 1
print(num)
arr = random.rand(5) # 1D array of 5 floats (between 0 and 1)
print(arr)
arr = random.rand(2, 3) # 2x3 array 
print(arr)
```

Random integers:

```python
from numpy import random

num = random.randint(10) # Generate a random integer number between 0 and 10
print(num)
num = random.randint(1, 10) # Generate a random integer number between 1 and 10
print(num)
arr = random.randint(1, 10, size=5) # 1D array of 5 integers between 1 and 10
print(arr)
arr = random.randint(1, 10, size=(2,3)) # 2x3 array 
print(arr)
```

choice()- allows us to generate a random value based on an array of values.

```python
from numpy import random

num = random.choice([3, 5, 7, 9])
print(num)
arr = random.choice([3, 5, 7, 9], size=(3, 5))
print(arr)
```
