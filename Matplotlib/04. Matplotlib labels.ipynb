Axis labels: xlabel() and ylabel() functions to set a label for the x- and y-axis.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

Title: title() function to set a title for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

Font Properties for Title and Labels: The 'fontdict' parameter in xlabel(), ylabel(), and title() to set font properties.

Common Font Properties:

family:	'serif', 'sans-serif', 'monospace' 

size:	12, 18, 'large', 'x-large'	

color:	'red', '#4CAF50'	Text color

weight:	'bold', 'normal', 'light'

style:	'italic', 'normal'

stretch:  'condensed', 'expanded'

variant:  'small-caps'

fontname:	'Times New Roman'

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

Position the title: The 'loc' parameter in title() to position the title. 
Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun"])
y = np.array([120, 150, 90, 180, 200, 160])

plt.plot(x, y, marker = '.', linestyle = 'dotted', color = 'black', markersize = 10, markerfacecolor = 'yellow')
plt.title("Monthly Sales (in RM)", fontdict = {'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': 20})
plt.xlabel("Months")
plt.ylabel("Sales (RM)")
plt.show()
```
