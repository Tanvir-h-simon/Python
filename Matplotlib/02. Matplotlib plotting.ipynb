plot(): The plot() function is used to draw points in a diagram.

```python
import matplotlib.pyplot as plt

x = [1, 8]
y = [3, 10]

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 8])
y_points = np.array([3, 10])

plt.plot(x_points, y_points)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points)
plt.show()
```

Default X-Points: Matplotlib needs two things to draw a line: X-values, Y-values. If we give both, it uses both. If we give only X-values, it must decide which is X and which is Y.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
plt.plot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10, 5, 7]) 

plt.plot(y_points) # x_points auto-generated
plt.show()
```

Plotting without line: To plot only the markers, we can use shortcut string notation parameter 'o', which means 'rings'.

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
