Marker(): The keyword argument marker to emphasize each point with a specified marker.

Marker codes:

'o'	circle

'.'	point

','	pixel

's'	square

'^'	triangle up

'v'	triangle down

'*'	star

'x'	x

'+'	plus

'D'	diamond

'p'	pentagon

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points, marker = 'o')
plt.show()
```

```python
plt.plot(x_points, y_points, marker = '.')
plt.show()
```

```python
plt.plot(x_points, y_points, marker = '*')
plt.show()
```

```python
plt.plot(x_points, y_points, marker = 'x')
plt.show()
```

Line styles:

'-'	solid(default)

'--'	dashed

'-.'	dashdot

':'	dotted

' ' None

Color codes:

b: blue

g: green

r: red

c: cyan

m: magenta

y: yellow

k: black

w: white

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points, marker = 'o', linestyle = 'dashed', color = 'black')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points, marker = 'o', ls = 'dashdot', c = 'black')
plt.show()
```

Marker Size:

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points, marker = 'o', linestyle = ':', color = 'b', markersize = 10)
plt.show()
```

```python
plt.plot(x_points, y_points, 'o:b', ms = 10) # The argument 'markersize' shorter version -> ms 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10])

plt.plot(y_points, marker='.', color='g', linestyle=':', markersize = 15)
plt.show()
```

```python
plt.plot(y_points, '.g:', ms = 15) 
plt.show()
```

Marker Color:

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10])

plt.plot(y_points, marker = 'o', linestyle = 'dotted', color = 'b', markersize = 10, markeredgecolor = 'r')
plt.show()
```

```python
plt.plot(y_points, 'o:b', ms = 10, mec = 'r') # The keyword argument 'markeredgecolor' shorter -> 'mec'
plt.show()
```

Marker Face Color:

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10])

plt.plot(y_points, marker = '.', color = 'b', linestyle = 'dotted', markersize = 20, markeredgecolor = 'b', markerfacecolor = 'g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10])
plt.plot(y_points, '.r:', ms = 20, mec = 'b', mfc = 'g') # The keyword argument 'markerfacecolor' shorter -> mfc
plt.show()
```

Hexadecimal color values:

Syntax: #rrggbb

Example,

rgb(255, 0 , 0)

#ff0000

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10])
plt.plot(y_points, marker = '.', linestyle = 'dashed', markersize = 20, markeredgecolor = '#4CAF50', markerfacecolor = '#4CAF50')
plt.show()
```

Line Width: 

```python
import matplotlib.pyplot as plt
import numpy as np

y_points = np.array([3, 8, 1, 10])

plt.plot(y_points, marker = 'o', color = 'black', markersize = 20, markerfacecolor = 'red', markeredgecolor = 'blue', linewidth = '10')
plt.show()
```

Multiple Lines:

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, marker = 'o', color = 'black', markersize = 20, markerfacecolor = 'black', markeredgecolor = 'yellow', linewidth = '5')
plt.plot(y2, marker = 'o', color = 'black', markersize = 20, markerfacecolor = 'black', markeredgecolor = 'yellow', linewidth = '5')

plt.show()
```
