### Generator

```python
lst = [x for x in range (500)] # Let, Dataset
# print(lst)

def data_loader(chunk_size, lst):
    for i in range(len(lst)):
        yield lst[i:i+chunk_size]

x = data_loader(5, lst)
# print(type(x))
# print(x)
print(next(x))
print(next(x))
```

### Lamda Function

```python
def square_addition(x, y):
    return x**2 + y**2

sq_add = lambda x, y: x**2 + y**2
print(sq_add(2, 3))    
```

```python
is_even = lambda x : x % 2 == 0

print(is_even(11))
```

### Filter

```python
numbers = [x for x in range (100)]
# print(numbers)
even = list(filter(lambda x : x % 2 == 0, numbers))
# print(even)
greater_fifty = list(filter(lambda x : x > 50, numbers))
print(greater_fifty)
```

### Reduce Function (Functools)

```python
from functools import reduce

lst = [1, 2, 3, 4, 5]

summation = reduce(lambda x, y : x + y, lst)
print(summation)
```

```python
max_value = reduce(lambda x, y : x if x > y else y, lst)
print(max_value)
```
