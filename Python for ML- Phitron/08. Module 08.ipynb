Class and Objects

```python
# class
class Phone:
  category = "Electronics"

  # constructor
  def __init__(self, model, battery, camera, battery_percentage = 100):
    self.model = model
    self.battery = battery
    self.camera = camera
    self.battery_percentage = battery_percentage

  # methods
  def charge(self, hour):
    self.charge += hour

  def capture(self, total_photo):
    if (self.battery_percentage) <= 0:
      print("No charge")
    else:
      self.battery_percentage -= total_photo
      print(f"Photo captured in {self.model}")

# objects
apple = Phone("Iphone17", 3000, 45)
motorola = Phone("B17", 4500, 50)
nokia = Phone("M17", 5500, 50)

print(apple.category, apple.model, apple.battery, apple.camera)
print(motorola.category, motorola.model, motorola.battery, motorola.camera)
print(nokia.category, nokia.model, nokia.battery, nokia.camera)

apple.capture(10)
apple.battery_percentage
```

Polymorphism

```python
class Camera:
  def __init__(self, name):
    self.name = name

  def capture(self):
    print("A photo is captured")
```

```python
class Smart_Phone(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution

  # Method overriding
  def capture(self):
    print("Photo is captured by Smart Phone")

class DSLR(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution

  # Method overriding
  def capture(self):
    print("Photo is captured by DSLR")

class Mirrorless(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution

  # Method overriding
  def capture(self):
    print("Photo is captured by Mirrorless Camera")

phone = Smart_Phone("Phone", 30)
dslr = DSLR("DSLR", 22.5)
mirrorless = Mirrorless("Mirrorless", 25)

phone.capture()
dslr.capture()
mirrorless.capture()
```

Encapsulation

```python
class Mobile:
  def __init__ (self, name, price, imei):
    self.name = name
    self.__price = price # Private
    self.__imei = imei # Private

  def get_price(self):
    return self.__price

  def get_imei(self):
    return self.__imei

my_iphone_encapsulated = Mobile("Iphone 17", 7000, "fgckxisf4364")

print(my_iphone_encapsulated.name)
# print(my_iphone_encapsulated.price)
# print(my_iphone_encapsulated.imei)
print(my_iphone_encapsulated.get_price())
print(my_iphone_encapsulated.get_imei())
```

Abstraction

```python
from abc import ABC, abstractmethod
class Telephone(ABC):
  @abstractmethod
  def make_call(self):
    pass

  @abstractmethod
  def capture_photo(self):
    pass

class Android(Telephone):
  def make_call(self):
    print("Making a call using android")

  def capture_photo(self):
    print("Capturing a photo using android camera")

class Iphone(Telephone):
  def make_call(self):
    print("Making a call using iphone")

  def capture_photo(self):
    print("Capturing a photo using iphone camera")
```

```python
android_phone = Android()
iphone_phone = Iphone()

android_phone.make_call()
android_phone.capture_photo()

iphone_phone.make_call()
iphone_phone.capture_photo()
```
