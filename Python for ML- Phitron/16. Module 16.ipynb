```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_df = pd.read_csv("sns_data.csv")
student_df
```

Line plot using Seaborn:

```python
sns.lineplot(data = student_df, x = 'week', y = 'attendance_rate')
```

```python
sns.lineplot(data = student_df, x = 'week', y = 'attendance_rate', errorbar = None, hue = 'gender')
```

```python
sns.relplot(kind = 'line', data = student_df, x = 'week', y = 'attendance_rate', errorbar = None, hue = 'gender')
```

```python
sns.relplot(kind = 'line', data = student_df, x = 'week', y = 'attendance_rate', errorbar = None, hue = 'class_level')
```

Scatter plot using seaborn: 

```python
sns.scatterplot(data = student_df, x = "study_hours", y = "test_score", hue = "gender", style = "class_level")
```

```python
sns.relplot(data = student_df, x = "study_hours", y = "test_score", hue = "gender", style = "class_level")
```

```python
tips_data = sns.load_dataset("tips")
tips_data
```

```python
sns.relplot(kind = 'scatter', data = tips_data, x = "total_bill", y = "tip", hue = "sex", style = "time")
plt.show()
```

Facet plot using seaborn:

```python
student_df = pd.read_csv("student_dataset_complete.csv")
student_df
```

```python
sns.relplot(kind= "scatter", data = student_df, x = "study_hours", y = "test_score", col="gender", row="hostel")
plt.show()
```

```python
sns.relplot(kind= "scatter", data = student_df, x = "study_hours", y = "test_score", col="week")
plt.show()
```

Histogram using seaborn:

```python
# Axis level
sns.histplot(data = student_df, x = "attendance_rate", hue="gender")
plt.show()
```

```python
sns.histplot(data = student_df, x = "attendance_rate", hue="gender", bins = 10, element="step")
plt.show()
```

```python
# Figure level
sns.displot(kind = "hist", data = student_df, x = "attendance_rate", col="gender")
plt.show()
```

```python
sns.displot(kind = "hist", data = student_df, x = "study_hours", col="gender")
plt.show()
```

KDE plot using seaborn:

```python
sns.kdeplot(data=student_df, x='attendance_rate')
plt.show()
```

```python
sns.displot(kind='hist', data=tips_data, x='total_bill', bins=20)
plt.show()
```

```python
sns.kdeplot(data = tips_data, x='tip')
plt.show()
```

```python
sns.kdeplot(data = tips_data, x='tip', hue='sex', fill=True)
plt.show()
```

Count plot and Bar plot using seaborn:

```python
sns.countplot(data=student_df, x = 'subject', hue='gender')
plt.show()
```

Bar plot:

```python
sns.barplot(data = student_df, x = "gender", y = "Marks")
plt.show()
```

```python
sns.barplot(data = student_df, x = "gender", y = "Marks", errorbar=None, estimator=np.median)
plt.show()
```

Reg and Lm Plot:

```python
# Axes level
sns.regplot(data = student_df, x = "study_hours", y = "test_score")
plt.show()
```

```python
# Figure level
sns.lmplot(data = student_df, x = "study_hours", y = "test_score", hue="gender")
plt.show()
```

Pair plot and Joint Plot:

```python
student_marks = student_df[["Marks", "study_hours", "attendance_rate", "gender"]]
student_marks
```

```python
sns.pairplot(data=student_marks)
plt.show()
```

```python
sns.pairplot(data=student_marks, kind="hist")
plt.show()
```

```python
sns.pairplot(data=student_marks, hue="gender")
plt.show()
```

```python
sns.jointplot(data=student_marks, x = "study_hours", y = "Marks")
plt.show()
```

```python
sns.jointplot(data=student_marks, x = "study_hours", y = "Marks", hue='gender')
plt.show()
```

```python
sns.jointplot(data=student_marks, x = "study_hours", y = "Marks", kind='kde')
plt.show()
```
