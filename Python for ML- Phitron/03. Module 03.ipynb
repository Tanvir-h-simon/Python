### Spliting, Joining

```python
prompt = "What is your name?"
tokens = prompt.split()
print(tokens)


prompt2 = "Hello,Hi,Hello"
tokens2 = prompt2.split(",")
print(tokens2)

# print(type(tokens2))

sentence = "-".join(tokens2)
print(sentence)

# print(type(sentence))
```

### Formatting

```python
name = "Tanvir"
age = 23
height = 3.1416
print(f"His name is {name}. He is {age} years old. He is {height} feet tall.\n")
```

### List

```python
numbers = [10, 20, 30, 40, 50]
# print(type(numbers))
print(numbers[1])

new_list = numbers[0 : 3]
print(new_list)

numbers.append(60);
print(numbers)

numbers.insert(1,5)
print(numbers)

numbers.remove(5)
print(numbers)

numbers.pop()
print(numbers)

numbers.sort()
print(numbers)

numbers.sort(reverse = True)
print(numbers)
```
