Handling duplicate values:

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah", "Alice"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New York", "Houston", "New York"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000, 50000]
}

df = pd.DataFrame(data)

df
```

```python
df.duplicated()

df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.drop_duplicates(subset=['City'], inplace=True)
df
```

Apply function on dataFrame

```python
minimum = df["Total Marks"].min()
maximum = df["Total Marks"].max()

df["Scaled Marks"] = df["Total Marks"].apply(lambda x: (x - minimum) / (maximum - minimum))
df
```

```python
def grading_system(marks):
    if marks >= 260:
        return "A+"
    elif marks >= 250:
        return "A"
    else:
        return "A-" 
    
df["Grade"] = df["Total Marks"].apply(grading_system)

df  
```

Datetime and Time Delta:

```python
import pandas as pd

df = pd.read_csv("student_data_with_datetime.csv")

df
```

```python
df['EnrollmentDate'] # dtype: object
```

```python
df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'])
df['EnrollmentDate'] # dtype: datetime
```

```python
df["FinishedDate"] = pd.to_datetime(df["FinishedDate"])
df["FinishedDate"]
```

```python
df['Enrollment Year'] = df['EnrollmentDate'].dt.year
df
```

```python
df['Enrollment Date'] = df['EnrollmentDate'].dt.day
df
```

```python
df['Total time taken to finish'] = df["FinishedDate"] - df['EnrollmentDate']
df
```
