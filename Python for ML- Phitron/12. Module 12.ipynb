Transforming to df from inbuilt data:

```python
this_list = [['Alice', 25], ['Bob', 30], ['Charlie', 28]]
list_df = pd.DataFrame(this_list)
print(list_df)
print(type(list_df))
```

```python
this_list = [['Alice', 25], ['Bob', 30], ['Charlie', 28]]
list_df = pd.DataFrame(this_list, columns=['Name', 'Age'], index=[1, 2, 3])
print(list_df)
```

```python
this_tuple = (('Alice', 25), ('Bob', 30), ('Charlie', 28))
tuple_df = pd.DataFrame(this_tuple, columns=['Name', 'Age'], index=[1, 2, 3])
print(tuple_df)
```

```python
this_dict = {
            'Name' : ['Alice', 'Bob','Charlie'],
            'Age' : [25, 28, 27]
}

dict_df = pd.DataFrame(this_dict, index=[1, 2, 3])
print(dict_df)
```

Sorting dataframe:

```python
df.sort_values('Data Structure Marks', inplace=True)

df
```

```python
df.sort_values('Data Structure Marks', ascending=False, inplace=True)

df
```

```python
df.sort_values(['Data Structure Marks', 'Python Marks'], ascending=[0,1], inplace=True) # 0 = False, 1= True

df
```

Filtering data:

```python
not_started = df.loc[df['CompletionStatus'] == 'Not Started']

not_started
```

```python
completed = df.loc[df['CompletionStatus'] == 'Completed']

completed
```

```python
completed = df.loc[(df['CompletionStatus'] == 'Completed') & (df['Data Structure Marks'] >= 90)]

completed
```
